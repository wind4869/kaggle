```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame

train_set = pd.read_csv('train.csv')
train_set
```

```python
train_set.info()
```

```python
train_set.describe()
```

```python
import matplotlib
import matplotlib.pyplot as plt

plt.style.use('ggplot')

fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(25, 10))

train_set['Survived'].value_counts().plot.bar(ax=axes[0,0])
train_set['Pclass'].value_counts().plot.bar(ax=axes[0,1])
train_set['Embarked'].value_counts().plot.bar(ax=axes[0,2])
train_set.plot.scatter(x='Survived', y='Age', ax=axes[1,0])

train_set['Age'][train_set['Pclass'] == 1].plot.kde(ax=axes[1,1]) 
train_set['Age'][train_set['Pclass'] == 2].plot.kde(ax=axes[1,1])
train_set['Age'][train_set['Pclass'] == 3].plot.kde(ax=axes[1,1])

plt.show()
```
